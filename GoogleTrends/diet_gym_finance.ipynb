```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('./data/multiTimeline.csv', skiprows=1)
df.head()
```

```python
df.info()
```

```python
df.columns = ['month', 'diet', 'gym', 'finance']
df.month = pd.to_datetime(df.month)
#change index of series to time (month)
df.set_index('month', inplace=True)
df.head()
```

```python
df.plot()
```

```python
df.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Take rolling average to smooth out trends

```python
diet = df[['diet']]
ax = diet.rolling(10).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, label = 'Rolling = 10')
diet.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, ax = ax,  label = 'Rolling = 12')
plt.xlabel('Year', fontsize=20);
```

```python
diet = df[['gym']]
ax = diet.rolling(6).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, label = 'Rolling = 10')
diet.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, ax = ax,  label = 'Rolling = 12')
plt.xlabel('Year', fontsize=20);
```
