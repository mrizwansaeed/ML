```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('./data/multiTimeline_dgse.csv', skiprows=1) #, usecols=[0,1,2,4])
df.head()
```

```python
df.info()
```

```python
df.columns = ['month', 'diet', 'gym', 'salad', 'exercise']
df.month = pd.to_datetime(df.month)
#change index of series to time (month)
df.set_index('month', inplace=True)
df.head()
```

```python
df.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Take rolling average to smooth out trends i.e., removing **seasonality**

```python
diet = df[['diet']]
ax = diet.rolling(10).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, label = 'Rolling = 10')
diet.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, ax = ax,  label = 'Rolling = 12')
plt.xlabel('Year', fontsize=20);
```

```python
gym = df[['gym']]
ax = gym.rolling(10).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, label = 'Rolling = 10')
gym.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, ax = ax,  label = 'Rolling = 12')
plt.xlabel('Year', fontsize=20);
```

Value of 12 for smoothing average provides a smoother curve.

```python
salad = df[['salad']]
ax = salad.rolling(10).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, label = 'Rolling = 10')
salad.rolling(12).mean().plot(figsize=(20,10), linewidth=5, fontsize=20, ax = ax,  label = 'Rolling = 12')
plt.xlabel('Year', fontsize=20);
```

Using *rolling average* to remove seasonality

```python
df_rm = pd.concat([diet.rolling(12).mean(), gym.rolling(12).mean(), salad.rolling(12).mean(), df[['exercise']].rolling(12).mean()], axis=1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
print(df_rm.corr()) #correlation after removing seasonlity
```

Season pattern in time series data (differencing)

```python
salad.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Correlations in raw data

```python
df.corr()
```

Salad and Gym seem to be strongly correlated in long term trends.

```python
df_rm.corr() #seasonality removed, keeping long term trends
```

```python
#stripping down to seasonality
df[['gym', 'diet', 'exercise']].diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Diet, gym, and exercise have highly correlated seasonal trends.

```python
df.diff().corr()
```
